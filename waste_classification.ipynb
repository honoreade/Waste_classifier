## Loading Tensorflow Pre-trained model

```python
# import os
# os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"

import tensorflow
import numpy as np
import matplotlib.pyplot as plt
from keras.preprocessing import image
from tensorflow.keras.utils import load_img, img_to_array
from PIL import Image
import io

# Declare some parameters
labels = ['Cardboard','Glass','Metal','paper','Plastic','Trash']
picturePath = r"C:\Users\honor\Downloads\images (6).jpg"

model = tensorflow.keras.models.load_model("Garbage.h5", compile=False)
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['acc'])
```

```python
# test_image = load_img(picturePath, target_size = (img_width, img_height)) 
img = load_img(picturePath, target_size=(64, 64))
img = img_to_array(img, dtype=np.uint8)
img = np.array(img)/255.0

p=model.predict(img[np.newaxis, ...])
```

```python
#print("Predicted shape",p.shape)
print("Maximum Probability: ",np.max(p[0], axis=-1))
predicted_class = labels[np.argmax(p[0], axis=-1)]
print("Classified:",predicted_class)
```
